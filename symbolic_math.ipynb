```python

import tokenizer as to
def power_token(input):
    input = to.parse(input)
    result = []
    i = 0
    while (i < len(input)):
        if(input[i].type == 'A'):
            if(i + 1 < len(input) and input[i+1].value == '^'):
                token = to.token('X', '')
                token.variable = input[i].value
                token.power = input[i + 2].value
                result.append(token)
                i += 3
            else:
                token = to.token('X', '')
                token.variable = input[i].value
                token.power = 1
                result.append(token)
                i+=1
        else:
            if(input[i].value != '*'):
                result.append(input[i])
            i += 1
    return result

def split(input):
    input = power_token(input)
    result = []
    bucket = []
    def commit():
        nonlocal bucket, result
        result.append(bucket)
        bucket = []
    for i in input:
        if(i.value == '+'):
            commit()
        else:
            bucket.append(i)
    commit()
    return result

def operate(input):
    input = split(input)
    result = []
    for i in input:
        current = {}
        numeric = 1
        zero = False
        for j in i:
            if(j.type == 'X'):
                if(j.variable in current.keys()):
                    current[j.variable].power += j.power
                else:
                    current[j.variable] = j 
            elif(j.type == '0'):
                if(j.value == 0):
                    zero = True
                    break
                numeric *= j.value
        if(not zero):
            line = []
            if(numeric != 0):
                line.append(to.token('0', numeric))
            result.append(line + [current[i] for i in current.keys()])
    

    def compare(a, b):
        for i in a[1:]:
            found = False
            for j in b[1:]:
                if(i == j):
                    found = True
                    break
            if not found:
                return False

        return True

    i = 0
    while(i < len(result)):
        j = i+1
        while(j < len(result)):
            if(compare(result[i], result[j])):
                result[i][0].value += result[j][0].value
                del result[j]
            else:
                j += 1
        i += 1
    return result

output = operate('x^4 * y + x^4 + 10 2x x^3')

print(output)
```
